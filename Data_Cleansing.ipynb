```python
dt['Alley'].fillna("No_Access", inplace=True)
dt['MasVnrType'].fillna("None", inplace=True)
dt['BsmtQual'].fillna("No_Basement", inplace=True)
dt['BsmtCond'].fillna("No_Basement", inplace=True)
dt['BsmtExposure'].fillna("No_Basement", inplace=True)
dt['BsmtFinType1'].fillna("No_Basement", inplace=True)
dt['BsmtFinType2'].fillna("No_Basement", inplace=True)
dt['FireplaceQu'].fillna("No_Fireplace", inplace=True)
dt['GarageType'].fillna("No_Garage", inplace=True)
dt['GarageFinish'].fillna("No_Garage", inplace=True)
dt['GarageQual'].fillna("No_Garage", inplace=True)
dt['GarageCond'].fillna("No_Garage", inplace=True)
dt['PoolQC'].fillna("No_Pool", inplace=True)
dt['Fence'].fillna("No_Fence", inplace=True)
dt['MiscFeature'].fillna("None", inplace=True)
```
