```python
dt.isnull().sum()
```

```python
miss = ['LotFrontage', 'Alley', 'MasVnrType', 'MasVnrArea', 'BsmtQual',\
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2','FireplaceQu', 'GarageType',\
       'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC', 'Fence',\
       'MiscFeature']
```

```python
def miss_data(data, col):
    for item in col:
        if (data[item].dtypes == 'object'):
            data[item].fillna("None", inplace=True)
        elif (data[item].dtypes == 'int64'):
            data[item].fillna(0, inplace=True)
        else:
            data[item].fillna(0.00, inplace=True)
```

```python
miss_data(dt, miss)
```

# Feature Selection

```python
train_X = dt.iloc[:,:-1]
train_X.head()
```

```python
train_y = dt['SalePrice']
train_y.head()
```

```python
train_X['BuildAge'] = train_X['YrSold']-train_X['YearBuilt']
train_X['RemodAge'] = train_X['YrSold']-train_X['YearRemodAdd']
train_X['GarageAge'] = train_X['YrSold']-train_X['GarageYrBlt']
train_X.head()
```

```python
train_X.isna().sum()
```

```python
train_X['GarageAge'].fillna(0, inplace=True)
train_X['GarageAge']=train_X['GarageAge'].astype('int64')
train_X['GarageAge'].isna().sum()
```

```python
train_X.drop(['YearBuilt', 'YearRemodAdd', 'GarageYrBlt'],axis=1,inplace=True)
```

```python
train_X.head()
```

```python
qual = ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond','PoolQC']
```

```python
def qual_ordinal(data, col):
    category = {
    'None':0, 
    'Po':1, 
    'Fa':2, 
    'TA':3, 
    'Gd':4, 
    'Ex':5}
    
    for item in col:
        data[item].replace(category, inplace=True) 
```
