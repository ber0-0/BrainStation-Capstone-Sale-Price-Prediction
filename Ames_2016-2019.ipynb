```python
ames = df_ames[['GeoRefNo','ClassPr_S','ClassSc_S','MA_Line1',\
                'MA_City', 'MA_State', 'MA_Zip1', 'PA-Strt','LndAc_S','ImpAc_S',\
                'OthAc_S','TtlVal_AsrYr', 'Neighborhood', 'LotArea', 'ParType',\
                'BldgNo_S', 'DwlgNo_S', 'BldgType', 'YrBuilt', 'HouseStyle',\
                'Foundation', 'RoofMatl', 'Ext1', 'Ext2', 'MasVnrType', 'Heating',\
                'Central Air', 'GLA', 'TtlBsmtSF', 'TotRmsAbvGrd', 'NmbrBRs','Fireplaces',\
                'PoolArea', 'GarageType', 'GarYrBlt', 'Cars', 'GarageArea',\
                'YrSold_YYYY', 'MoSold_MM', 'SalePrice', 'SaleType', 'SaleCond']]
```

```python
ames.shape
```

```python
ames.dtypes
```

```python
obj = list(ames.select_dtypes(include='object').columns)
obj
```

```python
for element in obj:
    print(f'{element}:{ames[element].unique()}')
```

```python
ames1 = ames[(ames['ClassSc_S'] == '1-STORY 1946 & NEWER ALL STYLES')|\
             (ames['ClassSc_S'] == '1-STORY 1945 & OLDER')|\
             (ames['ClassSc_S'] == '1-STORY TWNHM - 1946 & NEWER')|\
             (ames['ClassSc_S'] == '1-STORY W/FINISHED ATTIC ALL AGES')|\
             (ames['ClassSc_S'] == '1-1/2 STORY - UNFINISHED ALL AGES')|\
             (ames['ClassSc_S'] == '1-1/2 STORY FINISHED ALL AGES')|\
             (ames['ClassSc_S'] == '1-1/2 STORY TWNHM - ALL AGES')|\
             (ames['ClassSc_S'] == '2 FAMILY CONVERSION - ALL STYLES AND AGES')|\
             (ames['ClassSc_S'] == '2-1/2 STORY ALL AGES')|\
             (ames['ClassSc_S'] == '2-STORY 1945 & OLDER')|\
             (ames['ClassSc_S'] == '2-STORY 1946 & NEWER')|\
             (ames['ClassSc_S'] == '2-STORY TWNHM - 1946 & NEWER')|\
             (ames['ClassSc_S'] == 'APARTMENTS')|\
             (ames['ClassSc_S'] == 'CONDO')|\
             (ames['ClassSc_S'] == 'DUPLEX - ALL STYLES AND AGES')|\
             (ames['ClassSc_S'] == 'LAND')|\
             (ames['ClassSc_S'] == 'SPLIT FOYER')|\
             (ames['ClassSc_S'] == 'SPLIT OR MULTI-LEVEL')|\
             (ames['ClassSc_S'] == 'MOBILE HOME')]
```

```python
ames1.shape
```

```python
for element in obj:
    print(f'{element}:{ames1[element].unique()}')
```

```python
ames1.drop(['BldgType','HouseStyle', 'MA_Line1', 'MA_City', 'MA_State'], axis=1, inplace=True)
```

```python
ames1.shape
```

```python
ames1.isna().sum()
```

```python
ames1['SaleType'].fillna('None', inplace=True)
ames1['SaleType'].isna().sum()
```

```python
ames1['SaleCond'].fillna('None',inplace=True)
ames1['SaleCond'].isna().sum()
```

```python
ames1.dropna(subset=['Neighborhood'],inplace=True)
ames1.isna().sum()
```

```python
ames1.shape
```

```python
ames_transaction = ames1[ames1['SalePrice'].notnull()]
```

```python
ames_transaction.drop_duplicates(inplace=True)
ames_transaction.shape
```

```python
ames_transaction.head()
```

```python
ames1['Neighborhood_RecentSoldPrice'] = ames1.groupby(['Neighborhood','ClassSc_S'])['SalePrice'].transform(lambda x: x.fillna(x.mean()))
ames1['Neighborhood_RecentSoldPrice'] = ames1.groupby(['Neighborhood','ClassSc_S'])['TtlVal_AsrYr'].transform(lambda x: x.fillna(x.mean()))
```

```python
ames1.isna().sum()
```

```python
ames1['YrSold_YYYY'].fillna('None',inplace=True)
ames1['MoSold_MM'].fillna('None',inplace=True)
```

```python
ames1.isna().sum()
```

```python
ames_assessment = ames1.drop(['SalePrice','SaleType','SaleCond'], axis=1)
```

```python
ames_assessment.drop_duplicates(inplace=True)
ames_assessment.shape
```
